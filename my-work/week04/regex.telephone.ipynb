```python
import pandas as pd

csv='./people-100.csv'
data = pd.read_csv(csv)

pattern = r'(\(|\+1-|001-|)(\d{3})-?.?(\d{3})-?.?(\d{4})x?(\d*)?'
replace_phone = r'+1\2\3\4 '
replace_extension = r'\5'
data['Phone Cleaned'] = data['Phone'].str.replace(pat=pattern, repl=replace_phone, regex=True)
data['Extension'] = data['Phone'].str.replace(pat=pattern, repl=replace_extension, regex=True)
data
```
