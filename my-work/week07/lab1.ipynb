```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://cli.fusio.net/cli/climate_data/webdata/mly4935.csv"
data = pd.read_csv(url, skiprows=19)
data.head()
```

```python
corrtemp = data['month'].corr(data['meant'])
corrtemp
```

```python
df = data[['month', 'wdsp']]
df.dropna(inplace=True)
df.head()
```

```python
df['wdsp'] = df.loc[:, ('wdsp')].replace(' ', np.nan).astype(float)
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
corrwind = df['month'].corr(df['wdsp'])
corrwind
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.lmplot(x='month', y='wdsp', data=df, order=3, aspect=1.5, scatter_kws={'alpha':0.5})
plt.show()
```
